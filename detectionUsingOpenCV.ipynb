**Face Detection with OpenCV using Python**


**Import Library**

```python
import cv2
```

**Using the pre-trained classifier from OpenCV**

```python
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_alt.xml') #Load the cascade
```

**Upload the input image**

```python
img = cv2.imread("/content/sample1.jpg")
```

**Perform face detection**

```python
faces = face_cascade.detectMultiScale(img, 1.1, 4)
```

**Drawing Boxes for Faces**

```python
for(x, y, w, h) in faces:
  cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

Exporting the image file

```python
cv2.imwrite("processedimage.jpg", img)
print('Sucsessful')
```
